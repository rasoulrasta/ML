**Goal:** Given $X, y, \theta$, compute $J(\theta)$.

```python
#necessery library and pakage
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

```

$$J(\theta) = \frac{1}{2} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) ^ 2 = \frac{1}{2} \sum_{i=1}^{m} (\theta^T x^{(i)} - y^{(i)}) ^ 2 = \frac{1}{2} (X \theta - y)^T (X \theta - y)$$

```python
def sse(X,Y,teta):
    """ Sum of Squared Error Cost Function
    A vectorized implementation of the Sum of Squared Error.
    """
    error=X @ teta-Y
    
    return (0.5*error.T @ error)[0,0]
```

```python
#  Input data for Linear Regression 
# X is a array m*2 as the sample data (m samples of x0,x1)
X= np.reshape(np.array(list(map(int, input("enter m*2 number as Sample data(array):").split())),dtype=np.int8),(-1,2))
# Y is a array m*1 as the label 
Y=np.array(list(map(int, input("enter m number as Label (array):").split())),dtype=np.int8).reshape((-1,1))
#teta is a array 2*1 as coefficient (teta0,teat1)
teta=np.array(list(map(int, input("enter 2 number as Coefficient(array):").split())),dtype=np.int8).reshape((2,1))
```

```python
#definion of plot function
def hypo_plot(X,Y,teta):
    """ the inputs are 3 arrays as the sample data, label and coefficient nad out put draw hypothesis function
    """
    plt.scatter(X[:, 1], Y, s=100, c='r', marker='x')
    
    x = np.linspace(0.8, 3.2)
    x = x[:, None]
    x = np.hstack((np.ones_like(x), x))
    # plot hypothesis
    h = x @ teta
    plt.plot(x[:, 1], h, c='k', lw=2)
    plt.show()

    
```

```python
hypo_plot(X,Y,teta)
sse(X,Y,teta)
```

```python
j=[]
teta_s=np.linspace(-1,3,num=50)
for teta_1 in teta_s:
    cost=sse(X,Y,np.array([[0],[teta_1]]))
    j+=[cost]
```

```python
print(j)
```

```python
plt.plot(teta_s, j)
plt.plot(1, 0, 'ro')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$J(\theta)$')

plt.annotate('Global Optimum',
             xy=(1, 1), xycoords='data',
             xytext=(-15, 50), textcoords='offset points',
             arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center', verticalalignment='bottom')

plt.grid(True)
plt.show()
```
